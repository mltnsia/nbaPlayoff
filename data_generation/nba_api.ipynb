```python
# list of seasons to pull data from API
seasons = ["2005-06", "2006-07", "2007-08", "2008-09", "2009-10", "2010-11", "2011-12", "2012-13", "2013-14",
           "2014-15", "2015-16", "2016-17", "2017-18", "2018-19", "2019-20", "2020-21", "2021-22", "2022-23"]

# initialize df with latest full season 21/22
df = endpoints.leaguestandings.LeagueStandings(season="2004-05").standings.get_data_frame()[['SeasonID', 'TeamID', 'TeamCity', 'TeamName', 'Conference', 'PlayoffRank', 'WinPCT', 
                                                                                             'DiffPointsPG','DivisionRank','LeagueRank','LongHomeStreak','LongRoadStreak','LongWinStreak',
                                                                                             'LongLossStreak','ConferenceGamesBack','DivisionGamesBack','ClinchedConferenceTitle',
                                                                                             'ClinchedDivisionTitle','EliminatedConference','EliminatedDivision']]

# concatenate all seasons in list to df
for i in seasons:
    temp_df = endpoints.leaguestandings.LeagueStandings(season=i).standings.get_data_frame()[['SeasonID', 'TeamID', 'TeamCity', 'TeamName', 'Conference', 'PlayoffRank', 'WinPCT', 
                                                                                              'DiffPointsPG','DivisionRank','LeagueRank','LongHomeStreak','LongRoadStreak','LongWinStreak',
                                                                                              'LongLossStreak','ConferenceGamesBack','DivisionGamesBack','ClinchedConferenceTitle',
                                                                                              'ClinchedDivisionTitle','EliminatedConference','EliminatedDivision']]
    df = pd.concat([df, temp_df])
```

```python
# change seasonid type to numeric
df["SeasonID"] = df["SeasonID"].str[1:]
df["SeasonID"] = pd.to_numeric(df["SeasonID"])
```

```python
# playoffrank lag lists that will become columns
PlayoffRankLag1 = []
PlayoffRankLag2 = []
PlayoffRankLag3 = []
PlayoffRankLag4 = []

# winpct lag lists that will become columns
WinPCTLag1 = []
WinPCTLag2 = []
WinPCTLag3 = []
WinPCTLag4 = []

# diffpointspg lag lists that will become columns
DiffPointsPGLag1 = []
DiffPointsPGLag2 = []
DiffPointsPGLag3 = []
DiffPointsPGLag4 = []

DivisionRankLag1 = []
DivisionRankLag2 = []
DivisionRankLag3 = []
DivisionRankLag4 = []

LeagueRankLag1 = []
LeagueRankLag2 = []
LeagueRankLag3 = []
LeagueRankLag4 = []

LongHomeStreakLag1 = []
LongHomeStreakLag2 = []
LongHomeStreakLag3 = []
LongHomeStreakLag4 = []

LongRoadStreakLag1 = []
LongRoadStreakLag2 = []
LongRoadStreakLag3 = []
LongRoadStreakLag4 = []

LongWinStreakLag1 = []
LongWinStreakLag2 = []
LongWinStreakLag3 = []
LongWinStreakLag4 = []

LongLossStreakLag1 = []
LongLossStreakLag2 = []
LongLossStreakLag3 = []
LongLossStreakLag4 = []

ConferenceGamesBackLag1 = []
ConferenceGamesBackLag2 = []
ConferenceGamesBackLag3 = []
ConferenceGamesBackLag4 = []

DivisionGamesBackLag1 = []
DivisionGamesBackLag2 = []
DivisionGamesBackLag3 = []
DivisionGamesBackLag4 = []

ClinchedConferenceTitleLag1 = []
ClinchedConferenceTitleLag2 = []
ClinchedConferenceTitleLag3 = []
ClinchedConferenceTitleLag4 = []

ClinchedDivisionTitleLag1 = []
ClinchedDivisionTitleLag2 = []
ClinchedDivisionTitleLag3 = []
ClinchedDivisionTitleLag4 = []

EliminatedConferenceLag1 = []
EliminatedConferenceLag2 = []
EliminatedConferenceLag3 = []
EliminatedConferenceLag4 = []

EliminatedDivisionLag1 = []
EliminatedDivisionLag2 = []
EliminatedDivisionLag3 = []
EliminatedDivisionLag4 = []

# 2004/05 was the first season NBA had 30 teams
# to get 3 lagged years for each team, final season has to be 2007/08

# loop through df and append 3 lags into lists
for i, row in df.iterrows():
    currentTeamID = row.TeamID
    currentSeasonID = row.SeasonID
    if currentSeasonID >= 2008:
        for n in range(1, 5):
            eval(f'PlayoffRankLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["PlayoffRank"]) # append lagged playoff rank into list
            eval(f'WinPCTLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["WinPCT"]) # append lagged winpct into list
            eval(f'DiffPointsPGLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["DiffPointsPG"]) # append lagged diffpointspg into list
            eval(f'DivisionRankLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["DivisionRank"])
            eval(f'LeagueRankLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["LeagueRank"])
            eval(f'LongHomeStreakLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["LongHomeStreak"])
            eval(f'LongRoadStreakLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["LongRoadStreak"])
            eval(f'LongWinStreakLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["LongWinStreak"])
            eval(f'LongLossStreakLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["LongLossStreak"])
            eval(f'ConferenceGamesBackLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["ConferenceGamesBack"])
            eval(f'DivisionGamesBackLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["DivisionGamesBack"])
            eval(f'ClinchedConferenceTitleLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["ClinchedConferenceTitle"])
            eval(f'ClinchedDivisionTitleLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["ClinchedDivisionTitle"])
            eval(f'EliminatedConferenceLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["EliminatedConference"])
            eval(f'EliminatedDivisionLag{n}').append(df[(df["SeasonID"] == currentSeasonID - n) & (df["TeamID"] == currentTeamID)].iloc[0]["EliminatedDivision"])

    else:
        for n in range(1, 5):
            eval(f'PlayoffRankLag{n}').append(None)
            eval(f'WinPCTLag{n}').append(None)
            eval(f'DiffPointsPGLag{n}').append(None)
            eval(f'DivisionRankLag{n}').append(None)
            eval(f'LeagueRankLag{n}').append(None)
            eval(f'LongHomeStreakLag{n}').append(None)
            eval(f'LongRoadStreakLag{n}').append(None)
            eval(f'LongWinStreakLag{n}').append(None)
            eval(f'LongLossStreakLag{n}').append(None)
            eval(f'ConferenceGamesBackLag{n}').append(None)
            eval(f'DivisionGamesBackLag{n}').append(None)
            eval(f'ClinchedConferenceTitleLag{n}').append(None)
            eval(f'ClinchedDivisionTitleLag{n}').append(None)
            eval(f'EliminatedConferenceLag{n}').append(None)
            eval(f'EliminatedDivisionLag{n}').append(None)

# convert lagged lists into df columns
for n in range(1, 5):
    df[f"PlayoffRankLag{n}"] = eval(f'PlayoffRankLag{n}')
    df[f"WinPCTLag{n}"] = eval(f'WinPCTLag{n}')
    df[f"DiffPointsPGLag{n}"] = eval(f'DiffPointsPGLag{n}')
    df[f"DivisionRankLag{n}"] = eval(f'DivisionRankLag{n}')
    df[f"LeagueRankLag{n}"] = eval(f'LeagueRankLag{n}')
    df[f"LongHomeStreakLag{n}"] = eval(f'LongHomeStreakLag{n}')
    df[f"LongRoadStreakLag{n}"] = eval(f'LongRoadStreakLag{n}')
    df[f"LongWinStreakLag{n}"] = eval(f'LongWinStreakLag{n}')
    df[f"LongLossStreakLag{n}"] = eval(f'LongLossStreakLag{n}')
    df[f"ConferenceGamesBackLag{n}"] = eval(f'ConferenceGamesBackLag{n}')
    df[f"DivisionGamesBackLag{n}"] = eval(f'DivisionGamesBackLag{n}')
    df[f"ClinchedConferenceTitleLag{n}"] = eval(f'ClinchedConferenceTitleLag{n}')
    df[f"ClinchedDivisionTitleLag{n}"] = eval(f'ClinchedDivisionTitleLag{n}')
    df[f"EliminatedConferenceLag{n}"] = eval(f'EliminatedConferenceLag{n}')
    df[f"EliminatedDivisionLag{n}"] = eval(f'EliminatedDivisionLag{n}')

# remove rows with nan values
df = df[df['PlayoffRankLag1'].notna()]

# create target column from playoff rank
df['isPlayoff'] = (df['PlayoffRank'] <= 8).astype(int)

# split df into east and west conferences
# df_east = df[df["Conference"] == "East"]
# df_west = df[df["Conference"] == "West"]

# drop current playoff rank, winpct, diffpointspg, and conference columns
# df_east = df_east.drop(columns=['PlayoffRank', 'Conference', 'WinPCT', 'DiffPointsPG', 'TeamCity', 'TeamName']).reset_index(drop=True)
# df_west = df_west.drop(columns=['PlayoffRank', 'Conference', 'WinPCT', 'DiffPointsPG', 'TeamCity', 'TeamName']).reset_index(drop=True)
df = df.drop(columns=['PlayoffRank', 'WinPCT', 'DiffPointsPG', 'TeamCity', 'TeamName', 'DivisionRank','LeagueRank','LongHomeStreak','LongRoadStreak','LongWinStreak',
                                                                                             'LongLossStreak','ConferenceGamesBack','DivisionGamesBack','ClinchedConferenceTitle',
                                                                                             'ClinchedDivisionTitle','EliminatedConference','EliminatedDivision']).reset_index(drop=True)
```
